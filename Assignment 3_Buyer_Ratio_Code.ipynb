## H0- All proportions are equal
## Ha-  Not all Proportions are equal

```python
import pandas as pd
import scipy
from scipy import stats
from scipy.stats import chi2
```

```python
data=pd.read_csv('BuyerRatio.csv')
```

```python
data
```

```python
data_1=data.iloc[:,1:6]
```

## Using Chi-Square Test

```python
Chi_Square= scipy.stats.chi2_contingency(data_1)
Chi_Square[0]
```

```python
Chi_Square[1]
```

```python
Expected_result=Chi_Square[3]
```

```python
Chi_Square_result=sum([(o-e)**2/e for o,e in zip(data_1.values,Expected_result)])
Chi_Square_statistics= Chi_Square_result[0]+Chi_Square_result[1]
pvalue=1-chi2.cdf(Chi_Square_statistics,3)
```

```python
if pvalue <= 0.05:
    print('Accept Ha')
else:
    print('Accept H0')
```
